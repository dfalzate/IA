```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('Automobile.csv')
df.shape
df.info()
```

```python
df.describe(include='all').T
```

```python
sb.histplot(data=df,x='price');
```

```python
plt.title('Histogram')
plt.xlim(3000,50000)
plt.ylim(0,70)
plt.xlabel('Price of cars')
plt.ylabel('Frequency')
sb.histplot(data=df,x='price', color='red');

```

```python
sb.histplot(data=df,x='price', color='red', bins=20);
```

```python
sb.histplot(data=df,x='price', color='red', binwidth=20);
```

```python
sb.histplot(data=df,x='price', color='red', binwidth=200);
```

```python
sb.histplot(data=df,x='price', color='red', kde=True);
```
